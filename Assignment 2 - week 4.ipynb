# 3.Bar graph of male vs female buyers participated in the sales

```python
# Analysing male and female buyers seperately
slice_carsdata= carsdata[['Buyer Gender']]
slice_carsdata
```

```python
# creating the dataset
data = {'Male':4948, 'Female':5052}
Buyers = list(data.keys())
Counts = list(data.values())
  
fig = plt.figure(figsize = (10,5))
# creating the bar plot
plt.bar(Buyers,Counts, color =['orange','green'],
        width = 0.2)
 
plt.xlabel("Male vs Female")
plt.ylabel("No. of Cars Purchased")
plt.title("Male vs Female Buyers participated in the sale")
plt.show()
```
